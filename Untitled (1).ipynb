```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("SampleSuperstore.csv")
df.head()
```

```python
region=df['Region'].value_counts().reset_index()
region.rename(columns={'index':'Region','Region':'Count'},inplace=True)
region
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Region',y='Count',data=region,color='cyan')
plt.title("Region wise analysis")
plt.show()
```

```python
sns.barplot(x='Region',y='Profit',data=df,color='red')
plt.show()
```

```python
sns.barplot(x='Region',y='Sales',data=df,color='blue')
plt.show()
```

# Central needs more focus because sales and profit are bit less than others. 

# EAST REGION ANALYSIS

```python
east=df[df['Region']=='East']
states=east['State'].unique()
states
```

```python
total_sale=[]
total_profit=[]
for i in states:
    df_state=df[df['State']==i]
    sales=df_state['Sales'].sum()
    profit=df_state['Profit'].sum()
    total_sale.append(sales)
    total_profit.append(profit)
total_sale    
total_profit
```

```python
east_region=pd.DataFrame({'e_state':states,'sale':total_sale,'profit':total_profit})
east_region
```

```python
east_region.sort_values(by=['sale'],ascending=False,inplace=True)
plt.figure(figsize=(20,20))
plt.subplot(2,1,1)
sns.set(font_scale=1.8)
sns.barplot(x='e_state',y='sale',data=east_region)
plt.title("Sale of east region")
plt.xticks(rotation=90)

plt.subplot(2,1,2)
sns.set(font_scale=1.8)
sns.barplot(x='e_state',y='profit',data=east_region)
plt.title("Profit of east region")
plt.xticks(rotation=90)
plt.show()
```

# New york is doing extremely well as they are doing amazing sales and getting good profit. On the other hand, pennsylvania and ohio did good sales as compare to others but both the states were in major loss as compare to others.

# CATEGORY ANALYSIS

```python
category=df.groupby('Category')['Quantity'].sum().reset_index()
category
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Category',y='Quantity',data=category,color='blue')
plt.show()
```

# Office supplies products are sold at a great amount overall in the United States. While furniture and technology products are almost similar.

```python
category_profits=df.groupby('Category')['Profit'].sum().reset_index()
category_sales=df.groupby('Category')['Sales'].sum().reset_index()
category_sales_profit=category_sales.merge(category_profits,on='Category')
category_sales_profit
```

```python
plt.figure(figsize=(14,6))
plt.subplot(1,2,1)
plt.bar(category_sales_profit['Category'],category_sales_profit['Sales'],width=0.6,color='green')
plt.bar(category_sales_profit['Category'],category_sales_profit['Profit'],width=0.6,color='red')

plt.subplot(1,2,2)
exp=[0.3,0,0]
plt.pie(category_sales_profit['Profit'],labels=category_sales_profit['Category'],explode=exp,shadow=True,autopct='%1.1f%%')
plt.legend(loc='best')
plt.title("Overall profit analysis")
plt.show()
```

# Technology is getting maximum profit followed by office supplies.
